## Function Definition

### Diarization

```python
def diarization_rttm(input_file,output_file,num_speakers):
  diarization = pipeline(input_file,num_speakers=num_speakers)
  with open(output_file, "w") as rttm:
      diarization.write_rttm(rttm)
```

### Batch Diarization
